```python
import seaborn as sns
import os
sns.set_style(style="darkgrid")
sns.__version__
```

```python
root_dir = os.path.dirname(os.getcwd())
data_dir = os.path.join(root_dir, 'data')
data_set_path = os.path.join(data_dir, 'tips.json')
```

```python
"""Read File"""
tips_data = sns.load_dataset('tips')
tips_data.head()
```

```python
tips_data.info()
```

```python
tips_data.value_counts(tips_data['time'])
```

```python
"""Scaler Plot"""
```

```python
"""
Target Variable: which column that we want to check
Input Variable: Compare columns

Hypo 1: tip increase with total_bill
Result: Almost True
"""
sns.relplot(
    data=tips_data,
    x='total_bill',
    y='tip'
)
```

```python
"""
Hypo 2: There may be some relation of tip based on day
Result: People like to tip more in Saturday/Sunday
"""
sns.relplot(
    data=tips_data,
    x="total_bill",
    y="tip",
    hue="day",
    
)
```

```python
"""
Hypo 2: People like to spend more on dinner than lunch
Result: True
"""
sns.relplot(
    data=tips_data,
    x="total_bill",
    y="tip",
    hue="day",
    col="time",
    style="sex",
)
```

```python
sns.relplot(
    data=tips_data,
    x="total_bill",
    y="tip",
    hue="day",
    col="time",
    style="sex",
    row="sex",
)
```
