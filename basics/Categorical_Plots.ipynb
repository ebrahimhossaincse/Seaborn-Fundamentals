```python
"""Data"""
penguins = sns.load_dataset('penguins')
tips = sns.load_dataset('tips')
diamonds = sns.load_dataset('diamonds')
titanic = sns.load_dataset('titanic')
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    jitter=False,
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    col='sex',
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    order=['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    hue='sex'
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='time',
    kind='swarm'
)
```
