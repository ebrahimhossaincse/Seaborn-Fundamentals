```python
import seaborn as sns
import os
import matplotlib.pyplot as plt
from click import style

sns.set_style(style="darkgrid")
sns.__version__
```

```python
"""
Seaborn’s relplot() function is a flexible interface for creating relational plots, which are used to visualize relationships between multiple variables in a dataset. It can create scatter plots and line plots depending on the type of data and the specification of the plot.
"""
```

```python
"""Data"""
tips = sns.load_dataset('tips')
tips.head()
```

```python
"""SCATTER PLOT EXAMPLES"""
```

```python
"""Basic Scatter Plot
Description: This is a simple scatter plot that visualizes the relationship between the total bill (total_bill) and the tip amount (tip). Each point represents one observation in the dataset. Scatter plots are useful for exploring correlations between two continuous variables.
"""
# Basic scatter plot
sns.relplot(
    data=tips,
    x="total_bill", 
    y="tip"
)
```

```python
""" Scatter Plot with Hue (Color by Category)
Description: Here, the points are colored according to the sex variable, which introduces a categorical distinction. By using hue, you can easily compare different groups (in this case, male and female customers) on the scatter plot. Different colors represent different categories.
"""
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)
```

```python
"""Scatter Plot with Style (Different Markers)
Description: In this plot, both color (hue="sex") and marker style (style="time") are used to distinguish the data. Points are colored based on gender (sex) and different marker shapes represent the time of the day (time). This is useful when you want to explore multiple categorical variables on the same scatter plot.
"""
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="time"
)
```

```python
"""Scatter Plot with Size (Variable Marker Size)
Description: This scatter plot introduces variable marker sizes using the size parameter. The marker size is proportional to the size of the party (size column), allowing an additional dimension of information to be conveyed. Larger groups of customers are represented by larger markers, making it easier to spot trends based on group size.
"""
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    style="size",
    size="size",
    sizes=(50, 200)
)
```

```python
"""Scatter Plot with Multiple Subplots (Using Col)
Description: This scatter plot creates two subplots, one for each time of the day (lunch and dinner). The col parameter creates a grid of plots by splitting the data into multiple categories. It’s a great way to compare relationships across different subsets of the data.
"""
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    col="time"
)
```

```python
"""Scatter Plot with Row and Col (Multiple Subplots by Two Categories)
Description: This plot uses both the col and row parameters to create a grid of scatter plots. In this case, it splits the data by both time (columns) and smoker (rows). This allows for even more granular comparisons across different combinations of variables.
"""
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    col="time",
    row="smoker"
)
```

```python
"""Scatter Plot with Different Aspect and Height
Description: This plot adjusts the aspect ratio (aspect=0.6) and height (height=5) of the subplots. Changing the aspect ratio and height can make the plots more readable, especially when dealing with wide or tall plots. It also helps in aligning your plots with other elements of your visualizations.
"""
sns.relplot(
    x="total_bill", 
    y="tip", 
    hue="sex", 
    col="time", 
    data=tips, 
    aspect=0.5, 
    height=6
)
```

```python
"""LINE PLOT EXAMPLES"""
fmri = sns.load_dataset("fmri")
```

```python
"""Line Plot (Default for Time Series Data)
Description: This example creates a line plot, which is useful for visualizing changes over time or another continuous variable. The data here shows changes in brain activity signals over time (timepoint). Line plots are great for trends, especially in time series data.
"""
sns.relplot(
    x="timepoint", 
    y="signal", 
    kind="line", 
    data=fmri
)
```

```python
"""Line Plot with Hue (Multiple Lines by Category)
Description: In this line plot, multiple lines are drawn, one for each category in the event column. Each line represents the signal change over time for a different event, with color distinctions. This allows you to compare trends for different groups in the same plot.
"""
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri)
```

```python
"""Line Plot with Style (Different Line Styles)
Description: Here, different line styles are used to differentiate between the event categories. This is useful when color alone isn’t sufficient, or if you want to create a plot that’s easily interpretable in grayscale. Each event gets its own line style (solid, dashed, etc.).
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    style="event",
    kind="line",
    
)
```

```python
"""Line Plot with Confidence Intervals (CI)
Description: This line plot includes a shaded area around the lines, representing the confidence interval (CI). Here, the ci="sd" parameter indicates that the shaded area represents the standard deviation around the mean signal for each event, providing insight into the variability of the data.
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    kind="line",
    ci="sd",
)
```

```python
"""Line Plot with Markers
Description: This line plot includes markers on each data point along the lines, making it easy to see the actual observations rather than just the smoothed lines. This is helpful when you want to highlight specific data points, in addition to showing the trend.
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    kind="line",
    markers=True,
)
```

```python
"""Line Plot with Dashes (Different Line Patterns)
Description: This plot disables dashed lines by setting dashes=False. All the lines will be solid, which can be useful if you want to keep the plot simple and consistent, especially when you are using other visual cues like color.
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    kind="line",
    dashes=False,
)
```

```python
"""Line Plot with Multiple Subplots (Using Col)
Description: This is a line plot with subplots, where each subplot represents a different subject in the subject column. You can easily see the trends for each subject, making it easier to compare the same relationship across multiple individuals.
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    col="subject",
    kind="line",
    col_wrap=4
)
```

```python
"""Facet Grid of Line Plots (Row and Col for Line Plot)
Description: This is a grid of line plots, where the data is split across both rows (by region) and columns (by subject). Each plot represents the signal data for one region and one subject, making it easy to explore how the signal varies across different subjects and regions.
"""
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    col="subject",
    kind="line",
    row="region",
    height=8,
    aspect=.25,
    linewidth=2.5
)
```

```python
"""Line Plot with Faceted Grids (For Multiple Categories)
Description: This example extends the grid layout with custom height and aspect ratios, ensuring that the plots have a good balance between readability and the number of subplots. It’s a useful technique when dealing with multi-level categorical variables.
"""
sns.relplot(
    data=fmri,
    x="timepoint", 
    y="signal", 
    hue="event", 
    col="subject", 
    kind="line", 
    height=5,
    aspect=.25,
    linewidth=2.5
)
```

```python
sns.relplot(
    data = fmri,x = 'timepoint', y = 'signal',
    kind = 'line',
    hue = 'region',
    style = 'event',
    dashes=False, 
    markers=True
)
```
